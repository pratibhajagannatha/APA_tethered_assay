### Reading in tethered data

```python
data_dir_tethered = '/tscc/nfs/home/pjaganna1/projects/apa_screen_yongsheng/analyzed_data/output_files_may22'
```

```python
all_table_tethered=pd.DataFrame()
with open(os.path.join(input_dir,'plate_row_list.txt')) as file:
    for line in file:
        for r in ['Downstream', 'Upstream']:
            x = line.strip('\n').split(' ')
            x.append(r)
            fileName = '_'.join(x) + '.txt'
            print('reading file : {}'.format(fileName))
            if os.stat(os.path.join(data_dir_tethered,fileName)).st_size != 0:
                t = pd.read_csv((os.path.join(data_dir_tethered,fileName)), sep='\t', index_col=0)
                t['RBPmean-NEGmean'] = t['mean']-t.loc['NEG','mean']
                t['|RBPmean-NEGmean|'] = abs(t['RBPmean-NEGmean'])
                t['mean_normalized'] = t['mean']/t.loc['NEG','mean']
                t['std_normalized'] = t['std']/t.loc['NEG','mean']

                if r=='Downstream':
                    t['reporter']='Downstream'#annotating whether it is upstream or downstream
                elif r=='Upstream':
                    t['reporter']='Upstream'#annotating whether it is upstream or downstream
                all_table_tethered=pd.concat([all_table_tethered, t])
```

```python
all_table_tethered = all_table_tethered.reset_index()
```

```python

def get_norm_values(df, exp, reporter):
    untethered_norm = []
    for r in ['CPSF5', 'CPSF6', 'HNRNPCL1']:
        for rep in ['rep1', 'rep2', 'rep3']:
            val = np.log2(df.set_index('index').loc[r, rep]/df.set_index('index').loc['NEG', rep])
            untethered_norm.append([r, exp, reporter, rep, val])
    return pd.DataFrame(untethered_norm, columns=['RBP','experiment','reporter','replicate', 'norm_value'])
```

```python
get_norm_values(untethered_down_controls, 'untethered', 'downstream')
```

```python
normalized_control_table=pd.concat([get_norm_values(tethered_up_controls, 'tethered', 'upstream'), 
           get_norm_values(tethered_down_controls, 'tethered', 'downstream'),
          get_norm_values(untethered_up_controls, 'untethered', 'upstream'),
          get_norm_values(untethered_down_controls, 'untethered', 'downstream')])
```
