```python
def return_fileName_list(df_path):
    df = pd.read_csv(df_path, sep='\t')
    #df['uniprot'] = df['index'].map(uniprot_gene_dict)
    df['fileName'] = df.apply(lambda x: f'{x["index"]}|{x["Accession"]}', axis=1)
    #df['fileName'] = df.apply(lambda x: f'{x["uniprot"]}', axis=1)
    return df.dropna()['fileName'].tolist()
```

```python
all_lists = {'downstream_activator' : downstream_activator, 'upstream_activator' : upstream_activator, 
             'downstream_inhibitor' : downstream_inhibitor, 'upstream_inhibitor' : upstream_inhibitor,
            'background' : background_list}
```

```python
## Generate commands to run hmmscan for domain detection
for l in all_lists:
    print(f"hmmscan -o {l}_hmmscan_out.txt --domtblout {l}_dom_hmmscan_out.txt --pfamtblout {l}_pfam_hmmscan_out.txt --noali --notextw Pfam-A.hmm {os.path.join(sequence_out, l)}_sequences.fasta")
```

```python
all_domain_analyses = {}
for l in all_lists:
    domain_hits = []
    with open(f'{l}_dom_hmmscan_out.txt', 'r') as file:
            for line in file:
                # Skip comment lines and empty lines
                if line.startswith("#") or not line.strip():
                    continue

                # Split the line on whitespace
                fields = line.split()
                #print(fields)

                # Check if the line corresponds to the domain table entries
                # Typically, the domain table lines start with a target name that does not start with '#'
                # Adjust the condition based on the specific hmmscan output format if necessary
                if len(fields) > 18 and not fields[0].startswith("#"):
                    # Create a dictionary for the domain hit
                    domain_hit = {
                        'target_name': fields[0],
                        'target_accession': fields[1],
                        'query_name': fields[3],
                        'query_accession': fields[4],
                        'full_sequence_evalue': float(fields[6]),
                        'full_sequence_score': float(fields[7]),
                        'full_sequence_bias': float(fields[8]),
                        'domain_number': int(fields[9]),
                        'of': int(fields[10]),
                        'domain_c_evalue': float(fields[11]),
                        'domain_i_evalue': float(fields[12]),
                        'domain_score': float(fields[13]),
                        'domain_bias': float(fields[14]),
                        'hmm_from': int(fields[15]),
                        'hmm_to': int(fields[16]),
                        'ali_from': int(fields[17]),
                        'ali_to': int(fields[18]),
                        'env_from': int(fields[19]),
                        'env_to': int(fields[20]),
                        'acc': float(fields[21]),
                        'target_description': ' '.join(fields[22:])
                    }
                    # Append the dictionary to the list of domain hits
                    domain_hits.append(domain_hit)

    # Convert list of domain hits to a DataFrame
    df = pd.DataFrame(domain_hits)
    
    all_domain_analyses[l] = df
```

```python
def domain_counts(domains):
    """ Count the occurrences of each domain in the list. """
    domain_count = {}
    for domain in domains:
        if domain in domain_count:
            domain_count[domain] += 1
        else:
            domain_count[domain] = 1
    return domain_count
```

```python
background_domain_counts = domain_counts(all_domain_analyses['background']['target_name'].tolist())
```

```python
all_domain_analyses['background'].columns
```

```python
counts_df = pd.DataFrame()
for l in all_lists:
    col_name = '_'.join(l.split('_')[0:2])                             
    category_domain_counts = domain_counts(all_domain_analyses[l]['target_name'].tolist())
    counts_df = pd.concat([counts_df,pd.DataFrame.from_dict(category_domain_counts, orient='index', columns=[col_name])], axis=1).replace(float('nan'), 0)
```

```python
from scipy.stats import fisher_exact
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
import numpy as np
```

```python
# Function to calculate Chi-squared test for each domain
def perform_chi_squared_test(row, group_name, background_sum, group_sum):
    # The observed counts for the domain
    observed = [row[group_name], row['background']]
    
    # The expected counts if there's no enrichment
    expected = [group_sum, background_sum]
    
    # Create the contingency table
    contingency_table = [observed, expected]
    
    # Perform the Chi-squared test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table, correction=False)
    
    # Return the p-value (or other statistics if needed)
    return p_value

# Apply the function to each row in the DataFrame
counts_df_chi_copy = counts_df.copy()
for l in list(counts_df.columns):
    if l != 'background':
        counts_df_chi_copy[f'p_value_{l}'] = counts_df_chi_copy.apply(perform_chi_squared_test, args=(l, counts_df['background'].sum(), counts_df[l].sum()), axis=1)
        counts_df_chi_copy[f'p_value_{l}_neg_log10'] = counts_df_chi_copy.apply(lambda x: -(np.log10(x[f'p_value_{l}'])), axis=1)
```

```python
## LOOKING FOR ENRICHMENT OF RBP-SPECIFIC DOMAINS
sns.clustermap(counts_df_chi_copy[counts_df_chi_copy.index.str.contains('|'.join(['RRM', 'DEAD', 'KH']))][[i for i in list(counts_df_chi_copy.columns) if '_neg_log10' in i]], 
               cmap='mako_r', 
               col_cluster=False, 
               figsize=(8, 15)
              )
# plt.savefig('protein_domain_enrichment_RBD_only.png')
# plt.savefig('protein_domain_enrichment_RBD_only.svg')
# plt.savefig('protein_domain_enrichment_RBD_only.jpg')
```

```python
keep_domains = []
for cat in ['downstream_activator', 'upstream_activator', 'downstream_inhibitor', 'upstream_inhibitor']:
    tmp = counts_df_chi_copy[[f'p_value_{cat}']].sort_values(by=f'p_value_{cat}', ascending=True)
    keep_domains = keep_domains + tmp[tmp[f'p_value_{cat}'] <0.05].head(5).index.tolist()
```

```python
## TOP 5 TERM ENRICHMENT
sns.clustermap(counts_df_chi_copy[counts_df_chi_copy.index.isin(keep_domains)][[i for i in list(counts_df_chi_copy.columns) if '_neg_log10' in i]], cmap='mako_r', figsize=(9, 10), col_cluster=False)
# plt.savefig('protein_domain_enrichment_top5.png')
# plt.savefig('protein_domain_enrichment_top5.svg')
# plt.savefig('protein_domain_enrichment_top5.jpg')
```
