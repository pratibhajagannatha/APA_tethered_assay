```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data_table = pd.read_csv('expt23_across_bio_reps.csv', index_col=0)
```

```python
data_table.head()
```

```python
data_table['x_label'] = data_table.apply(lambda x: f'{x["RBP+MCP"]}_{x["snRNA(+/-MS2)"]}_{x["PAS or no PAS"]}', axis=1)
```

```python
data_table['color_label'] = data_table.apply(lambda x: f'{x["snRNA(+/-MS2)"]}_{x["PAS or no PAS"]}', axis=1)
```

```python
data_table_filtered = data_table[(data_table['60bp or NT Spacer']=='60bp') & (data_table['Biological Replicate']!=1) ]
```

```python
plot_order=['FLAG_snRNA_NoPAS',
            'FLAG_MS2_snRNA_NoPAS',
            'FLAG_snRNA_PAS',
            'FLAG_MS2_snRNA_PAS',
            'CPSF6_snRNA_NoPAS',
            'CPSF6_MS2_snRNA_NoPAS',
            'CPSF6_snRNA_PAS',
            'CPSF6_MS2_snRNA_PAS',
            'RNPS1_snRNA_NoPAS',
            'RNPS1_MS2_snRNA_NoPAS',
            'RNPS1_snRNA_PAS',
            'RNPS1_MS2_snRNA_PAS'
           ]
```

```python
plt.figure(figsize=(4,4))
sns.barplot(data=data_table_filtered.sort_values('x_label'), 
            x='x_label', 
            y='Average of technical replicates',
           order=plot_order, hue='RBP+MCP',palette=['grey', 'steelblue', 'darkorange'],
           hue_order=['FLAG', 'CPSF6', 'RNPS1'])
# sns.stripplot(data=data_table_filtered.sort_values('x_label'), 
#             x='x_label', 
#             y='Average of technical replicates',
#            order=plot_order, color='black')
plt.xticks(rotation=45)
# plt.savefig('u7_component_testing.png')
# plt.savefig('u7_component_testing.jpg')
# plt.savefig('u7_component_testing.svg')
```

```python
stats_test_vals = []
for m in set(data_table_filtered['snRNA(+/-MS2)'].tolist()):
    for p in set(data_table_filtered['PAS or no PAS'].tolist()):
        for r in ['CPSF6', 'RNPS1']:
            print(r, m, p)
            rbp_vals = data_table_filtered[(data_table_filtered['RBP+MCP']==r) & (data_table_filtered['snRNA(+/-MS2)']==m) & (data_table_filtered['PAS or no PAS']==p)]['Average of technical replicates'].tolist()
            flag_vals = data_table_filtered[(data_table_filtered['RBP+MCP']=='FLAG') & (data_table_filtered['snRNA(+/-MS2)']==m) & (data_table_filtered['PAS or no PAS']==p)]['Average of technical replicates'].tolist()
            print(stats.ttest_ind(flag_vals, rbp_vals))
            stats_test_vals.append([r,m,p,stats.ttest_ind(flag_vals, rbp_vals)[1]])
stats_test_vals_df = pd.DataFrame(stats_test_vals, columns=['RBP+MCP','+/-MS2', 'PAS or no PAS','pval']).sort_values('RBP+MCP')
```

```python
def assign_stars(df):
    if df['pval'] <0.001:
        return '**'
    elif df['pval'] < 0.05:
        return '*'
    else:
        return ''
```

```python
stats_test_vals_df['stars']=stats_test_vals_df.apply(assign_stars, axis=1)
```
