```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
expt20 = pd.read_csv('expt20_across_bio_reps.csv', index_col=0)
```

```python
expt22 = pd.read_csv('expt22_across_bio_reps.csv', index_col=0)
```

```python
expt20 = expt20[expt20['Orientation of MS2 relative to spacer']=='MS2-Spacer-U7']
```

```python
sns.barplot(data=expt20, x='RBP+MCP', y='Average of technical replicates', hue='Guide Distance/NT')
```

```python
from scipy import stats
```

```python
distance_test_stats = []
for d in list(set(expt20['Guide Distance/NT'])):
    for r in ['CPSF5+MCP', 'CPSF6+MCP']:
        f = expt20[(expt20['Guide Distance/NT']==d) & (expt20['RBP+MCP'] =='FLAG+MCP')]['Average of technical replicates'].tolist()
        c5 = expt20[(expt20['Guide Distance/NT']==d) & (expt20['RBP+MCP'] ==r)]['Average of technical replicates'].tolist()
    
        distance_test_stats.append((r, d, stats.ttest_ind(f, c5)[1]))
```

```python
pd.DataFrame(distance_test_stats)
```

```python
fig, axes = plt.subplots(1, 5, sharey=True, figsize=(6,3))
a = 0
for d in ['0bp', '20bp', '40bp', '60bp', '120bp']:
    sns.barplot(data=expt20[expt20['Guide Distance/NT']==d].sort_values(by='RBP+MCP', ascending=False), x='RBP+MCP', y='Average of technical replicates',
               palette=['grey', 'cadetblue', 'steelblue'], hue_order=['FLAG+MCP'],order=['FLAG+MCP', 'CPSF5+MCP', 'CPSF6+MCP'], ax=axes[a], 
               )
    axes[a].set_title(d)
    axes[a].set_xticklabels(axes[a].get_xticklabels(), rotation=45)
    a+=1
plt.savefig('u7_distance_testing.png')
plt.savefig('u7_distance_testing.jpg')
plt.savefig('u7_distance_testing.svg')
```
